```python
##### import psycopg2

def run_system():
    try:
        # الاتصال بقاعدة البيانات - )
        conn = psycopg2.connect(
            dbname="GrandHorizonDB", 
            user="postgres", 
            password="123456", 
            host="localhost", 
            port="5432"
        )
        cursor = conn.cursor()
        print("✅ Connected to Grand Horizon Database!")
    except Exception as e:
        print(f"❌ Connection Error: {e}")
        return

    while True:
        print("\n" + "="*45)
        print("🏨 Grand Horizon Management System")
        print("="*45)
        print("1. View all active bookings")
        print("2. Add a new guest")
        print("3. Update guest email")
        print("4. Cancel a booking")
        print("5. Book a room (Stored Procedure)")
        print("6. Cancellation Report")
        print("7. Exit")

        choice = input("\nSelect an option (1-7): ")

        # 1. عرض الحجوزات النشطة
        if choice == '1':
            try:
                cursor.execute("SELECT * FROM bookings WHERE status = 'Active';")
                rows = cursor.fetchall()
                if not rows:
                    print("\n⚠️ No active bookings found.")
                else:
                    print("\n" + "="*75)
                    print(f"{'ID':<5} | {'Guest':<10} | {'Room':<10} | {'Check-in':<12} | {'Check-out':<12} | {'Status'}")
                    print("-" * 75)
                    for row in rows:
                        print(f"{row[0]:<5} | {row[1]:<10} | {row[2]:<10} | {str(row[3]):<12} | {str(row[4]):<12} | {row[6]}")
                    print("="*75)
            except Exception as e:
                print(f"❌ Error: {e}")
                conn.rollback()

        # 2. إضافة نزيل جديد (
        elif choice == '2':
            name = input("Enter Full Name: ")
            email = input("Enter Email: ")
            phone = input("Enter Phone: ")
            try:
                cursor.execute("INSERT INTO guests (full_name, email, phone) VALUES (%s, %s, %s);", (name, email, phone))
                conn.commit()
                print(f"✅ Guest '{name}' added successfully!")
            except Exception as e:
                print(f"❌ Error: {e}")
                conn.rollback()

        # 3. تحديث إيميل النزيل
        elif choice == '3':
            g_id = input("Enter Guest ID: ")
            new_mail = input("Enter New Email: ")
            try:
                cursor.execute("UPDATE guests SET email = %s WHERE guest_id = %s;", (new_mail, g_id))
                conn.commit()
                print("✅ Email updated.")
            except Exception as e:
                print(f"❌ Error: {e}")
                conn.rollback()

        # 4. إلغاء حجز
        elif choice == '4':
            b_id = input("Enter Booking ID to cancel: ")
            try:
                cursor.execute("UPDATE bookings SET status = 'Cancelled' WHERE booking_id = %s;", (b_id,))
                conn.commit()
                print("✅ Booking status updated to Cancelled.")
            except Exception as e:
                print(f"❌ Error: {e}")
                conn.rollback()

        # 5. حجز غرفة (استدعاء البروسيدجر)
        elif choice == '5':
            g_id = input("Guest ID: ")
            r_id = input("Room ID: ")
            in_date = input("Check-in (YYYY-MM-DD): ")
            out_date = input("Check-out (YYYY-MM-DD): ")
            try:
                cursor.execute("CALL make_booking(%s, %s, %s, %s);", (g_id, r_id, in_date, out_date))
                conn.commit()
                print("✅ Booking created via Procedure.")
            except Exception as e:
                print(f"❌ Error: {e}")
                conn.rollback()

        # 6. تقرير الإلغاءات
        elif choice == '6':
            try:
                cursor.execute("SELECT guest_id, COUNT(*) FROM bookings WHERE status = 'Cancelled' GROUP BY guest_id;")
                rows = cursor.fetchall()
                print("\n--- Cancellations Report ---")
                for row in rows:
                    print(f"Guest ID: {row[0]} | Total Cancellations: {row[1]}")
            except Exception as e:
                print(f"❌ Error: {e}")

        # 7. الخروج
        elif choice == '7':
            print("👋 Goodbye Raneem! System closing...")
            break

    cursor.close()
    conn.close()

# تشغيل الكود
run_system()
```
